```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=sns.load_dataset("car_crashes")
```

```python
dataset.head()
```

Scatterplot between total and speeding

```python
sns.scatterplot(x="total",y="speeding",data=dataset)
```

Relation plot for Total and speeding according to the alcohol consumption

```python
sns.relplot(x="total",y="speeding",data=dataset,hue="alcohol")
```

Scatterplot for Total and alcohol consumption

```python
sns.scatterplot(x="total",y="alcohol",data=dataset)
```

relationn between the insurance premium and the losses

```python
sns.scatterplot(x="ins_premium", y="ins_losses",data=dataset)
```

distance plot for speeding

```python
sns.histplot(dataset["speeding"])
```

relation between the not distracted people with people having first accident

```python
sns.scatterplot(x="not_distracted",y="no_previous",data=dataset)
```

Jointplot for Total and speeding

```python
sns.jointplot(x="total",y="speeding",data=dataset)
```
